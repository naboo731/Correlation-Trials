```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr
```

```python
absentee_data = pd.read_csv('Absenteeism_at_work.tsv', sep='\t')
absentee_data.head()
```

```python
absentee_data.columns
```

Correlation coefficient between **day of the week** and **reason for absence**

```python
absentee_data['Day of the week'].corr(absentee_data['Reason for absence'])
```

```python
pearsonr(absentee_data['Day of the week'], absentee_data['Reason for absence'])
```

```python
plt.figure(figsize=(12, 8))

plt.scatter(absentee_data['Day of the week'], 
            absentee_data['Reason for absence'], color='m')

plt.title('Day of the Week and Reason for Absence')

plt.xlabel('Day of the Week')
plt.ylabel('Reason for Absence')

plt.show()
```

```python
reasons_count = absentee_data.groupby('Reason for absence')['ID'].count().sort_values(ascending=False)
reasons_count
```

A: The correlation coefficient is 0.1163.. This demonstrates a positive correlation however it is not a strong positive correlation. The p-value of this correlation is 0.001526 which is a relatively large p-value demonstrating this correlation is **not** statistically significant.
